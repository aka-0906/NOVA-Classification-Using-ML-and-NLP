### Distribution of `novaclass` (Categorical Column)

Let's visualize the distribution of the `novaclass` column to better understand the observed imbalance.

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='novaclass', data=data, palette='viridis', hue='novaclass', legend=False)
plt.title('Distribution of Nova Classes')
plt.xlabel('Nova Class')
plt.ylabel('Count')
plt.show()
```

### Correlation Heatmap of Numerical Columns

To understand the relationships between the 106 numerical columns, a correlation heatmap can be very informative. We will use the `numerical_data` DataFrame to avoid issues with non-numeric types.

```python
plt.figure(figsize=(15, 12))
sns.heatmap(numerical_data.corr(), cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
numerical_data.corr()["novaclass"].sort_values(ascending=False)
```

### Exploring Categorical Columns: `catname` and `macroclass`

Let's examine the unique values and their counts for `catname` and `macroclass` to understand their distributions.

```python
print("Value counts for 'catname':")
display(data['catname'].value_counts())
```

```python
print("\nValue counts for 'macroclass':")
display(data['macroclass'].value_counts())
```

### Bar Plots for Categorical Column Distributions

Now, let's visualize the distributions of `catname` and `macroclass` using bar plots.

```python
plt.figure(figsize=(15, 8))
top_catnames = data['catname'].value_counts().head(20).index
sns.countplot(y='catname', data=data[data['catname'].isin(top_catnames)], order=top_catnames, palette='viridis', hue='catname', legend=False)
plt.title('Distribution of Top 20 Food Categories (catname)')
plt.xlabel('Count')
plt.ylabel('Category Name')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='macroclass', data=data, order=data['macroclass'].value_counts().index, palette='magma', hue='macroclass', legend=False)
plt.title('Distribution of Macro Classes')
plt.xlabel('Count')
plt.ylabel('Macro Class')
plt.tight_layout()
plt.show()
```
